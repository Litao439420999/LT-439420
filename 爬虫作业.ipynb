```python
#第二种方法

import requests
from bs4 import BeautifulSoup

url = 'http://www.gz.xinhuanet.com/'
headers = {
   'User-Agent':'zilla/5.0 (Windows NT 10.0; Win64; x64; rv:81.0) Gecko/20100101 Firefox/81.0' 
}

html = requests.get(url=url,headers=headers)
#print(html.encoding)
html.encoding = html.apparent_encoding
#print(html.encoding)

soup = BeautifulSoup(html.text,'lxml')
div_list = soup.find_all('div',class_="card_gzzw")
for div  in div_list:
    list = div.find_all('a')
    for i in list:
        title = i.text
        href = i['href']
        print('标题:'+title+'\t链接：'+href)
    
print('爬取结束！')
```

## 3 中国大学排名定向爬虫

中国大学排名：http://www.zuihaodaxue.cn/zuihaodaxuepaiming2019.html

```python
#第一种方法：
import requests
from bs4 import BeautifulSoup

headers = {
   'User-Agent':'zilla/5.0 (Windows NT 10.0; Win64; x64; rv:81.0) Gecko/20100101 Firefox/81.0' 
}
url="http://www.zuihaodaxue.cn/zuihaodaxuepaiming2019.html"

r=requests.get(url=url,headers=headers,timeout=30)
r.encoding = r.apparent_encoding

soup = BeautifulSoup(r.text,'lxml')
trlist = soup.select('.hidden_zhpm > tr ' )

with open('中国大学排名.txt','w+',encoding='utf-8')as f:
    for tr in trlist:
        tds = tr('td')
        no = tds[0].text
        name = tds[1].text
        address = tds[2].text
        score = tds[3].text
        print(no+'\t\t'+name+'\t\t'+address+'\t\t'+score)
        f.write(no+'\t\t'+name+'\t\t'+address+'\t\t'+score+'\n')
    
    
print("结束！")

```

```python
#第二种方法：
import requests
from bs4 import BeautifulSoup

headers = {
   'User-Agent':'zilla/5.0 (Windows NT 10.0; Win64; x64; rv:81.0) Gecko/20100101 Firefox/81.0' 
}
url="http://www.zuihaodaxue.cn/zuihaodaxuepaiming2019.html"

r=requests.get(url=url,headers=headers,timeout=30)
r.encoding = r.apparent_encoding

soup = BeautifulSoup(r.text,'lxml')

trlist = soup.find_all('tr',class_="alt")

for tr  in trlist:
    tdlist = tr.find_all('td')
    no = tdlist[0].text
    name = tdlist[1].text
    city = tdlist[2].text
    score = tdlist[3].text
    print(no+'\t\t'+name+'\t\t'+city+'\t\t'+score+'\n')
    
print('爬取结束！')
    

```
